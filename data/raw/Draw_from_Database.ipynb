## Pulling Data from AWS-S3

- The data being used can be found [here](https://www.kaggle.com/yamaerenay/spotify-dataset-19212020-160k-tracks).

```python
import boto3
# Pulling the data using built-in credentials
# https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/s3.html

s3 = boto3.resource("s3")
```

```python
dir(s3)
```

```python
print(s3.Bucket("spotify-data158").creation_date)
```

### Loading the data below

```python
import pandas as pd

track_obj = s3.Bucket("spotify-data158").Object("tracks.csv").get()
tracks = pd.read_csv(track_obj["Body"])

tracks.head()
```

```python
art_obj = s3.Bucket("spotify-data158").Object("artists.csv").get()
artists = pd.read_csv(art_obj["Body"])

artists.head()
```

```python
art_track = tracks.merge(artists, on=["id"])
art_track.head()
```

```python
data_obj = s3.Bucket("spotify-data158").Object("data_o.csv").get()
data = pd.read_csv(data_obj["Body"])

data.head()
```

```python
artdata_obj = s3.Bucket("spotify-data158").Object("data_by_artist_o.csv").get()
art_data = pd.read_csv(artdata_obj["Body"])

art_data.head()
```

```python
gendata_obj = s3.Bucket("spotify-data158").Object("data_by_genres_o.csv").get()
gen_data = pd.read_csv(gendata_obj["Body"])

gen_data.head()
```

```python
yeardata_obj = s3.Bucket("spotify-data158").Object("data_by_year_o.csv").get()
year_data = pd.read_csv(yeardata_obj["Body"])

year_data.head()
```
